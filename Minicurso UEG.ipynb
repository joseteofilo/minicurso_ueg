## Cálculo de Fingerprints (Morgan)

![alt text](https://drive.google.com/uc?id=1KTyXXHweCK4lHj9ee-Ch36pwXNOzEmOE)

```python
# filtra as colunas que queremos
moldf = moldf.filter(items=['is_active', 'SMILES',])
```

```python
# cria uma variável com as moléculas
mols = moldf["SMILES"]
```

```python
def rdkit_numpy_convert(fp):
  output = []
  for f in fp:
    arr = np.zeros((1,))
    DataStructs.ConvertToNumpyArray(f, arr)
    output.append(arr)
  return np.asarray(output)
```

```python
def smiles_to_fp(mols, radius, nbits):
  # convert smiles to RDKit mol object
  fps = []
  for mol in mols:
    mol = Chem.MolFromSmiles(mol, sanitize=True)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nbits)
    fps.append(fp)
  return fps
```

```python
# Adiciona uma colunas com fingerprints
moldf['fingerprints']= pd.DataFrame(smiles_to_fp(mols, 2, 2048))
# Uma coluna para cada bit
df_descriptors = pd.DataFrame(rdkit_numpy_convert(moldf['fingerprints']))
```

```python
df_descriptors
```

```python
x = df_descriptors.values
```

```python
len(x)
```

```python
len(y)
```

## Divisão dos dados em treino e teste

![alt text](https://drive.google.com/uc?id=1RzutpPjZwZb-vmU73RAKR0p9gFMlXMdI)

```python
# seleciona aleatoriamente 20% dos compostos como conjunto de teste
x_tr, x_ts, y_tr, y_ts = train_test_split(x, y, test_size=0.20, random_state=42, stratify=y)
```

## Geração do modelo de Random Forest - modelo simples

### Geração do modelo



```python
simple_rf = RandomForestClassifier(random_state=42, n_jobs = -1)

simple_rf.fit(x_tr, y_tr)
```

### Statistics

```python
# predict Outcome class
pred_rf = simple_rf.predict(x_ts)
```

```python
# Gera a matriz da confusão
confusion = metrics.confusion_matrix(y_ts, pred_rf)

TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]

# Gráfico da matriz da confusão
plt.figure(figsize=(5,5))
sns.heatmap(confusion, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Real');
plt.xlabel('Predito');
title = "Matriz da confusão"
plt.title(title, size = 15);
```

```python
# Calcula as métricas de validação
accuracy = accuracy_score(y_ts, pred_rf) #accuracy
kappa = cohen_kappa_score(y_ts, pred_rf) #kappa
sensitivity = metrics.recall_score(y_ts, pred_rf) #Sensitivity
specificity = TN / (TN + FP) #Specificity
precision = TP / float(TP + FP) #Precision
positive_pred_value = TP / float(TP + FP) #PPV
negative_pred_value = TN / float(TN + FN) #NPV


# Imprime as métricas de validação
print("Accuracy = ", accuracy)
print("Kappa = ", kappa)
print("Sensitivity = ", sensitivity)
print("Specificity = ", specificity)
print("NPV = ", negative_pred_value)
print("Precision = ", precision)
```

```python
# Gráfico de barras com as métricas de validação
sns.set_style("whitegrid")
stats = [accuracy, kappa, sensitivity, specificity, positive_pred_value, negative_pred_value]
labels = ["Acc", "Kappa", "Se", "Sp", "PPV", "NPV"]
report = sns.barplot(x=labels, y=stats)
```

```python
#converte as métricas de validação em um DataFrame do pandas
metrics_simple_rf = DataFrame({'Accuracy': accuracy, 'Kappa': kappa,
                    "Sensitivity": sensitivity,"Specificity": specificity,
                    "NPV": negative_pred_value,"Precision": precision,}, index=[0])

metrics_simple_rf
```

## Validação cruzada

![alt text](https://drive.google.com/uc?id=1hTqrtuihhw1MbXEXDd_gA3NOuauqYGbT)

```python
# Create folds for cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

### Construção do modelo

```python
# Make Gmean scorer
scorer = make_scorer(geometric_mean_score)

#Search for optimal tuning parameters and build the model
# create grid search dictionary
param_grid = {"n_estimators": [50, 100, 300], "max_depth": [60, 80], 'max_features': ['auto', 'sqrt'],
              'min_samples_leaf': [1, 2], 'min_samples_split': [2, 3, 5],}

opt_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=cv, scoring = scorer, verbose=0, n_jobs = -1)

opt_rf.fit(x_tr, y_tr)

print("val. score: %s" % opt_rf.best_score_)
print("test score: %s" % opt_rf.score(x_ts, y_ts))
print("Best params: %s" % opt_rf.best_params_)
```

### Statistics

```python
# Select best model in cross validation
rf = opt_rf.best_estimator_
```

```python
# predict Outcome class
pred_rf = rf.predict(x_ts)
```

```python
# Gera a matriz da confusão
confusion = metrics.confusion_matrix(y_ts, pred_rf)

TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]

# Gráfico da matriz da confusão
plt.figure(figsize=(5,5))
sns.heatmap(confusion, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Real');
plt.xlabel('Predito');
title = "Matriz da confusão"
plt.title(title, size = 15);
```

```python
# Calcula as métricas de validação
accuracy = accuracy_score(y_ts, pred_rf) #accuracy
kappa = cohen_kappa_score(y_ts, pred_rf) #kappa
sensitivity = metrics.recall_score(y_ts, pred_rf) #Sensitivity
specificity = TN / (TN + FP) #Specificity
precision = TP / float(TP + FP) #Precision
positive_pred_value = TP / float(TP + FP) #PPV
negative_pred_value = TN / float(TN + FN) #NPV

# Imprime as métricas de validação
print("Accuracy = ", accuracy)
print("Kappa = ", kappa)
print("Sensitivity = ", sensitivity)
print("Specificity = ", specificity)
print("NPV = ", negative_pred_value)
print("Precision = ", precision)
```

```python
# Gráfico de barras com as métricas de validação
sns.set_style("whitegrid")
stats = [accuracy, kappa, sensitivity, specificity, positive_pred_value, negative_pred_value]
labels = ["Acc", "Kappa", "Se", "Sp", "PPV", "NPV"]
report = sns.barplot(x=labels, y=stats)
```
